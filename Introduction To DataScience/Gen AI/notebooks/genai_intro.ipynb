## Theory### The AI Hierarchy```mermaidgraph TD    AI[Artificial Intelligence] --> ML[Machine Learning]    ML --> DL[Deep Learning]    DL --> GenAI[Generative AI]    style GenAI fill:#f9f,stroke:#333```**Artificial Intelligence (AI):** Broad field of creating intelligent systems.  **Machine Learning (ML):** Subset of AI — learning from data.  **Deep Learning (DL):** Subset of ML — using neural networks with many layers.  **Generative AI (GenAI):** Subset of DL — creating new data (text, images, audio, code).### How Generative AI Works1. **Training Phase:** Model learns patterns from a large corpus.2. **Inference Phase:** Model generates new content conditioned on a prompt.3. **Evaluation:** Generated content is assessed for quality, diversity, and faithfulness.
