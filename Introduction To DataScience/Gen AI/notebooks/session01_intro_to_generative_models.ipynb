## Theory### What is Generative AI?**Generative AI** refers to algorithms that can create new data samples — text, images, audio, code — that resemble the training distribution. Unlike **discriminative models** (which learn decision boundaries), generative models learn the joint probability distribution $P(X)$ or $P(X|Y)$ and can *sample* from it.### Key Families| Family | Core Idea | Example ||--------|-----------|--------|| Autoencoders / VAEs | Compress → latent space → reconstruct | Fashion-MNIST generation || GANs | Generator vs Discriminator game | StyleGAN, DeepFake || Diffusion Models | Iterative denoising from Gaussian noise | DALL·E 2, Stable Diffusion || Transformers (LLMs) | Autoregressive token prediction | GPT-4, LLaMA || Flow-based Models | Invertible transformations | Glow, RealNVP |### Architecture Diagram```mermaidgraph LR    A[Input Data] --> B[Encoder / Feature Extraction]    B --> C[Latent Space z]    C --> D[Decoder / Generator]    D --> E[Generated Output]    style C fill:#f9f,stroke:#333```
