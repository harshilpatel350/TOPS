```python
class atm:
    def __init__(self):
        self.pin = ''
        self.balance = 0
        self.card_activated = False
        self.card_activation()
    
    def card_activation(self):
        print("Welcome! Let's activate your ATM card.")
        name = input("Enter your full name for card registration: ")
        mobile = input("Enter your 10-digit mobile number: ")
        if len(mobile) == 10 and mobile.isdigit():
            print(f"Thank you {name}, your card has been activated.")
            self.card_activated = True
            self.create_pin()
        else:
            print("Invalid mobile number. Please try again.")
            self.card_activation()

    def create_pin(self):
        user_pin = input("Set your 4-digit ATM pin: ")
        if len(user_pin) == 4 and user_pin.isdigit():
            self.pin = user_pin
            print("Your pin has been successfully created.")
            self.menu()
        else:
            print("Invalid pin format. Please try again.")
            self.create_pin()

    def menu(self):
        user_input = input('''\n
        Hello, Greetings! How can I help you?
        1. Press 1 to change the pin
        2. Press 2 to deposit money
        3. Press 3 to check balance
        4. Press 4 to withdraw money
        5. Press 5 to exit
        Enter your choice: ''')

        if user_input == '1':
            self.change_pin()
        elif user_input == '2':
            self.deposit_money()
        elif user_input == '3':
            self.check_balance()
        elif user_input == '4':
            self.withdraw()
        elif user_input == '5':
            print("Thank you for using the ATM. Goodbye!")
            exit()
        else:
            print("Invalid input. Try again.")
            self.menu()

    def deposit_money(self):
        user_pin = input("Enter your pin: ")
        if user_pin == self.pin:
            try:
                amt = int(input("Enter the amount to deposit: "))
                if amt > 0:
                    self.balance += amt
                    print(f"₹{amt} has been deposited to your account.")
                else:
                    print("Deposit amount must be positive.")
            except ValueError:
                print("Invalid amount entered.")
            self.menu()
        else:
            print("Incorrect pin. Please try again.")
            self.deposit_money()

    def change_pin(self):
        user_pin = input("Enter your current pin: ")
        if user_pin == self.pin:
            new_pin = input("Enter the new 4-digit pin: ")
            if len(new_pin) == 4 and new_pin.isdigit():
                self.pin = new_pin
                print("Pin updated successfully.")
            else:
                print("Invalid pin format.")
        else:
            print("Incorrect pin.")
        self.menu()

    def withdraw(self):
        user_pin = input("Enter your pin: ")
        if user_pin == self.pin:
            try:
                amt = int(input("Enter amount to withdraw: "))
                charge = 0
                if amt <= 0:
                    print("Enter a valid amount.")
                elif amt <= self.balance:
                    if amt <= 1000:
                        charge = 0
                    elif amt <= 20000:
                        charge = 100
                    elif amt <= 100000:
                        charge = 1000
                    else:
                        charge = 2000

                    total = amt + charge
                    if total <= self.balance:
                        self.balance -= total
                        print(f"₹{amt} withdrawn. ₹{charge} charged as fee. Remaining balance: ₹{self.balance}")
                    else:
                        print("Insufficient balance including withdrawal charge.")
                else:
                    print("Insufficient balance.")
            except ValueError:
                print("Invalid input.")
        else:
            print("Incorrect pin.")
        self.menu()

    def check_balance(self):
        user_pin = input("Enter your pin: ")
        if user_pin == self.pin:
            print(f"Your account balance is ₹{self.balance}")
        else:
            print("Incorrect pin.")
        self.menu()

a = atm()
```
